# Sub plots

```python
x=np.linspace(1,15,10)
y=x**2
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g--o')
plt.subplot(2,2,2)
plt.plot(x,y,'b*-')
plt.subplot(2,2,3)
plt.plot(x,x,'r--')
plt.subplot(2,2,4)
plt.plot(x,y/2,'y>-')

# subplot is used to plot two or more graphs at once.
# identified and indexed based on indices
# (total rows, total columns, present index)
```

```python
# to plot graph inside a plot

fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])
#[left,bottom,width,height]

ax1.plot(x,y)

ax2.plot(y,x)
```

```python
import matplotlib.pyplot as plt
labels=["php","python","java","c"]
numbers=[75,70,80,60]
fig=plt.figure(facecolor='aquamarine',edgecolor='red')
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) # [left,bottom,width,height]
axes1.pie(numbers,labels=labels,autopct="%0.2f%%",colors=["r","g","pink","orange"])
axes1.legend()
plt.show()
```

```python
Explode=[0.2,0.05,0,0]
plt.pie(numbers,labels=labels,explode=Explode)
plt.show()
```
