```python
df=sns.load_dataset("tips")
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
```

inference

from the plot we can say that as the bill amount increses, the tip is also incresing,

which means they are directly proportional.

```python
sns.scatterplot(x="size",y="total_bill",data=df)
```

```python
sns.scatterplot(x="size",y="tip",data=df)
```

largest tip is given at the presence of 3 people

```python
sns.lineplot(x="size",y="total_bill",data=df,ci=None)

# ci means confidence interval
# if that is not given, then u will see a seperate interval also
```

```python
sns.distplot(df["tip"])
```

inference:

the tips are mostly ranged between 2 and 4

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="day") # rel plot is meant for relational plot
```

```python
df["day"].value_counts()
```

```python
sns.barplot(x="day",y="tip",data=df)
```

inference:

you can see the highest tips are on sundays

```python
sns.barplot(x="sex",y="tip",data=df,ci=None)
```

inference:

males give more tips than females
